## Indian Ethics

Primarily focusing on Hinduism, which is the third largest religion and widely regarded as the oldest religion in the world.

### **Hindu virtues**:
- **Aparigraha:** avoid hoarding and greediness
- **Brahmacharya**: consiously direct your energy
- **Asteya**: dont steal
- **Satya**: tell the truth
- **Ahimsa**: practice non-violence
- **Sauca**: cleanliness is important
- **Santosha**: try to be content
- **Tapas**: do your best
- **Svadhyaya**: study yourself
- **Ishvara Pranidhana**: stay humble

Virtues are something you cultivate through life. These are virtues that should be done unto others and to oneself.

---

### **Karma**
Karma is about actions and the consequences of those actions. It can be 
- **Fear-based**: *"You reap what you sow"*
- **Love-based**: *"Gifts will be given back to you"*

Karma in terms of individual actions, but all the totality of actions in ones life. Good actions weighing vs bad actions. The Hindu idea is reincarnation, but what you are reincarnated into depends on your cumulative lifes actions.

### **Mahabharata: Bhagavad Gita**

One of the two great Hindu ethics, one of the great epics of the world. Comparable to The Illiad and The Odyssey.

Key Point: 

&emsp;It is a war between cousins over who is the rightful heir to the throne. Symbolically, <br>&emsp;Arjuna is the one the fire the first arrow and start the war, but in this moment he <br>&emsp;faces interpersonal challenge, which Krishna talks him through.

Arjuna:

&emsp;**"O Krishna, if I must kill my kinsmen, I care not for victory, nor pleasure, nay not even for life itself."**

&emsp;What is the point of anything if we are going to war to kill loved ones?

Krishna:

&emsp;**"You will neither merit heaven, not achieve fame with such thoughts. Desist from this 
<br>&emsp;&nbsp;&nbsp;weakness and get ready to fight.**

&emsp;This reinforces the idea of **dharma**, or duty. It is Arjuna's duty to start this war.

Summary (chatGPT:)

    The Bhagavad Gita centers on a conversation between Prince Arjuna and his charioteer, Krishna, on the battlefield of Kurukshetra. Arjuna faces a moral dilemma about fighting in the war against his own family and teachers. Krishna advises him on his duty as a warrior to fight for righteousness (dharma), emphasizing selfless action, devotion to God, and the pursuit of spiritual wisdom, ultimately guiding Arjuna to act without attachment to the results.

*TLDR; It is duty that should drive action.*

## Classical Chinese Ethics

### **Confucianism**

Religion vs philosophy vs ethical tradition... depends on your cultural context.

<br>

**Dao de**
- ***Dao***: "way" - public, objective guidance
- ***De***: "virtue" - character traits, skills, and dispositions
- 'Way' comes first before 'virtue'. Public expectations should guide an individuals behavior
- Collectivist concept, not individualistic

<br>
 
**li vs. ren**
- ***li***: ritual: defined in terms of ones role in society
- ***ren***: humanity: universal moral sense
- Are we people? Or are we sisters, nephews, musicians, christian? How are we defined?
- Our roles in society vs. our fundamental humanity

<br>
 
Social relationships:
- Familial heirarchy
- Occupational heirarchy
- Social heirarchy

<br>

Confucian negative version of the Golden Rule:<br>
&emsp;&emsp;**"What you do not desire, do not effect on others"**

### **Zhuangzi**
- **shi vs. fei**
- *shi*: right (this)
- *fei*: wrong (not this)

*Standpoint epistemology*, or *situated knowledges*: The idea that you dont get truth by one global universal perspective that is from one spot and is omniscient. You get it by adding the different perspectives of everyone in the world. Positionality is very important.

You need to privilege the view from below. People who face the most adversity are the ones who see the world for what it is.

### **Confucian Traditions**

Confucian philosophy says you must understand your place in the heirarchy of the universe. Honor your elders and their elders and so forth. You must show respect to the dead, spending 3 years in ritual mourning on parents death, respecting the 3 years an infant spends completely dependent on parental care. 

Societal harmony is the key point.

Confucious viewed himself not as an original thinker, but of the product of his parents influence, and their parents influence, and the influence of society, and so forth.

## **Zhuangzi**


Zhuangzi once dreamt that he was a butterfly, joyfully fluttering about, not knowing he was Zhuangzi. Upon awakening, he found himself to be Zhuangzi again. However, he was uncertain if he was Zhuangzi who had dreamt of being a butterfly or a butterfly now dreaming of being Zhuangzi.

Whether the world we are living in is real or not has massive ramifications for our actions and what we may or may not do.
